# Scaling and Train testing

```python
x=x.dropna()
y=y[x.index]
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=42)
X_res,y_res=sm.fit_resample(x,y)
```

```python
p=X_res,y_res.value_counts().plot(kind='bar')
```

```python
# Standarlization
from sklearn.preprocessing import StandardScaler
scale_X=StandardScaler()


X_res= scale_X.fit_transform(X_res)    #standarlization
X_res =pd.DataFrame(X_res,columns=['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error', 'perimeter error', 'area error',
       'smoothness error', 'compactness error', 'concavity error',
       'concave points error', 'symmetry error', 'fractal dimension error',
       'worst radius', 'worst texture', 'worst perimeter', 'worst area',
       'worst smoothness', 'worst compactness', 'worst concavity',
       'worst concave points', 'worst symmetry', 'worst fractal dimension'])
X_res.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.20)
```

# DECISION TREE USING K FOLD CROSS VALIDATION TECHNIQUE

```python
#import necessary libaries
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
Decisiontree=DecisionTreeClassifier()

#specify the number of folds (k) for cross-validation
k=5

#create a KFold cross-validation object
kf=KFold(n_splits=k,shuffle=True,random_state=42)
```

```python
#performn k-fold cross-Validation and calculate the accuracy
scores=cross_val_score(Decisiontree,X_res,y_res,cv=kf,scoring='accuracy')

#print the  accuracy scores for each fold
for fold ,score in enumerate(scores):
    print(f'fold {fold +1}: acccuracy={score:.2f}')
    
#calculate and print the mean accuracy across all folds
mean_accuracy =np.mean(scores)
print(f"mean accuarcy ={mean_accuracy:.2f}")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import LeaveOneOut
from sklearn import datasets
from sklearn import svm

#create a support vector machine classifier
clf=svm.SVC(kernel='linear',C=1)

#perform leave-one -out cross-validation
loo=LeaveOneOut()
scores=cross_val_score(clf,X_res,y_res,cv=loo)

#print the accurcy for each fold
print("Accuracy for each fold:",scores)

#calculate and print the mean accuracy
print("mean accuracy:",scores.mean())
```
